Find the employee with the highest salary per department.\
Output the department name, employee's first name along with the corresponding salary.

```python
df_highest = employee.groupby(['department'])['salary'].max().reset_index().rename(columns = {'salary': 'highest_salary'})
df_highest
```

```python
# employee[df_highest['department'] == employee['department'] and df_highest['salary'] == employee['salary']]
```

```python
final_df = pd.merge(df_highest, employee, on = 'department', how = 'right')
#final_df
```

```python
final_df = final_df[final_df['salary'] == final_df['highest_salary']].drop('salary', axis = 1)#.rename(columns = {'salary_y': 'salary'})
# final_df = final_df.drop('salary_x', axis = 1).rename(columns = {'salary_y': 'salary'})
# final_df
```

```python
# Output the department name, employee's first name along with the corresponding salary.
final_df[['department', 'first_name', 'highest_salary']]
```

```python
# sql
with cte as (
select *, MAX(salary) OVER(PARTITION BY department) as highest_salary
from employee )

select department, first_name, salary
from cte
where salary = highest_salary;
```
