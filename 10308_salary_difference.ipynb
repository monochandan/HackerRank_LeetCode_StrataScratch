# Calculates the difference between the highest salaries in the marketing and engineering departments. Output just the absolute difference in salaries.

```python
# group by department, max salary , deferance
new_df = pd.merge(df_employee, df_departments, left_on = 'department_id', right_on = 'id', how = 'left')
```

```python
new_df
```

```python
abs(new_df[new_df['department'].isin(['marketing', 'engineering'])].groupby('department')['salary'].max().reset_index()['salary'].diff().iloc[1])
```

```python
# sql
with cte as (
select d.department, max(salary) max_salary
from df_employee e
left join df_department d
on e.department_id = d.id
where d.department in('engineering', 'marketing')
group by d.department )

select abs(max_salary - max_sala) as salary_difference
from 
(select max_salary, lead(max_salary) over() as max_sala
from cte) w
where w.max_sala is not null
```
