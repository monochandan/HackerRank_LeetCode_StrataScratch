## find the number of times the words "bull" and "bear" occur in the contents.We are counting the number of times the words occur so wordslike bullish should not be included in our count. output the word bull and bear along with the corresponding number of occurances

```python
data = {
    "filename": ['draft1.txt', 'draft2.txt', 'final.txt', 'text1.txt', 'text3.txt'],
    "content": [
        "The bull market is thriving, but not for bearish investors.",
        "A bear and a bull walked into a market. The bear triumphed.",
        "The bull ran through the market while the bear watched carefully.",
        "Market trends are confusing, but neither bull nor bear prevails today.",
        "No bull or bear in this flat market."
    ]
}

df = pd.DataFrame(data)
print(df)
```

# Answer

```python
## creare a dictionary
my_dict = {"bull": int(df['content'].str.count(r'\bbull\b', flags = re.I).sum()), 
           "bear": int(df['content'].str.count(r'\bbear\b', flags = re.I).sum())}
```

```python
my_dict
```

```python
## dictionary to dataframe
df = pd.DataFrame(list(my_dict.items()), columns = ['word', 'count'])
df
```

```python
df['bull'] = df['content'].apply(lambda text: len(re.findall(r'\bbull\b', text)))
df['bear'] = df['content'].apply(lambda text: len(re.findall(r'\bbear\b', text)))
```

```python
df[['bull', 'bear']].sum().reset_index()
```

```python
## count the words of each content and store the numeric value in  a seperate column called word_count
df['word_count'] = df['content'].apply(lambda text: len(text.split()))
```

# High-Density Areas ID 10544
## Identify the top 3 areas with the highest customer density. Customer density = (total number of unique customers in the area / area size).

## Your output should include the area name and its calculated customer density, and ties will be ranked the same.

```python
data = {
    "transaction_id": [1, 2, 3, 4, 5],
    "customer_id": [3, 7, 11, 11, 8],
    "transaction_amount": [189, 495, 207, 483, 290],
    "transaction_date": [
        "2024-01-21 00:00:00",
        "2024-01-29 00:00:00",
        "2024-02-17 00:00:00",
        "2024-01-08 00:00:00",
        "2024-02-14 00:00:00",
    ],
    "store_id": [4, 1, 5, 5, 2],
}

tr = pd.DataFrame(data)

tr.head()
```

```python
stores_data = {
    "store_id": [1, 2, 3, 4, 5],
    "store_location": ["City Center", "Suburbs", "Downtown", "Industrial Area", "Mall"],
    "store_open_date": ["2018-09-11", "2019-08-10", "2012-05-11", "2013-07-19", "2013-02-05"],
    "area_name": ["Area A", "Area B", "Area C", "Area D", "Area E"],
    "area_size": [121, 238, 70, 128, 160],
}

stores_df = pd.DataFrame(stores_data)
stores_df.head()
```

```python
print(tr.columns)
print(stores_df.columns)
```

```python
# identify the top 3 areas with the highest customre density
# customer density = total number of unique customres in the area / area size

## marge 2 data set:
df_marged = pd.merge(tr, stores_df, on = 'store_id')
df_marged = df_marged[['customer_id', 'area_name', 'area_size']]
df_marged
```

```python
df_marged['total_number_of_unique_customer'] = df_marged.groupby(df_marged['area_name'])['customer_id'].transform('nunique')
#pandas.core.series.Series
#total_number_of_unique_customer.values
```

```python
# uniq customre for the each area
df_marged['customer_density'] = df_marged['total_number_of_unique_customer'] / df_marged['area_size']
```

```python
df_marged = df_marged[['area_name', 'customer_density']].sort_values('customer_density', ascending = False).head(3)
#df_marged = df_marged[['area_name', 'customer_density']].head(3)
df_marged
```

You are working on a data analysis project at Deloitte where you need to analyze a dataset containing information
about various cities. Your task is to calculate the population density of these cities, rounded to the nearest integer, and identify the cities with the minimum and maximum densities.
The population density should be calculated as (Population / Area).


The output should contain 'city', 'country', 'density'.

```python
data = {
    "city": ["Metropolis", "Smallville", "Coastcity", "Starcity", "Gotham"],
    "country": ["Countryland", "Countryland", "Oceanland", "Mountainous", "Islander"],
    "population": [1000000, 50000, 300000, 600000, 1500000],
    "area": [500, 1000, 0, 600, 300]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df['density'] = df['population'] / df['area']
df['density'] = df['density'].replace([float('inf'), -float('inf')], 0)
```
