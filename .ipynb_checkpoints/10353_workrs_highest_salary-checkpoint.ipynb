Find the job titles of the employees with the highest salary. If multiple employees have the same highest salary, include the job titles for all such employees.

```python
df = pd.merge(worker, title, left_on = 'worker_id', right_on = 'worker_ref_id', how = 'left')
```

```python
df
```

```python
## boolean indexing
df[df['salary'] == df['salary'].max()][['worker_title']].rename(columns = {'worker_title': 'best_paid_title'})
```

```python
# sql
with cte as (
select *, max(salary) over() as max_salary
from worker w
left join title t
on w.worker_id = t.worker_ref_id )

select worker_title as bect_paid_title
from cte
wher salary = max_salary
```
