```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html

```python
# baisc graph
x, y = [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]


## resize the graph
plt.figure(figsize=(5,3), dpi=100) ## dpi possibly higher for the better print

#y = [2, 4, 6]
## keyword argument notation
# plt.plot(x, y, label = '2x', color = 'pink', linewidth = 2, linestyle = 'dashed', marker = '.', mec = 'black')

# shorthand notation
## formater fmt = '[color][marker][line]'

plt.plot(x, y, 'r.--', label = '2x') 

## adding title , specid√üfy the font, font size , font color in the fontdict parameter
plt.title('our first graph', fontdict = {'fontname': 'comic sans ms', 'fontsize': 11, 'color': 'green'})

x2 = np.arange(0,4,0.5)
# plot part of the graph as a line
plt.plot(x2[:5], x2[:5]**2, color = 'blue', label = 'x^2')
## plot remaining of the graph
plt.plot(x2[4:], x2[4:]**2, color = 'green', label = 'x^2', linestyle = 'dashed')

## adding xlabel and ylabel
plt.xlabel('x-axis', fontdict = {'color': 'red'})
plt.ylabel('y-axis', fontdict = {'color': 'blue'})

# if we want to have the integer as the x-axis and y-axis, define the xticks and yticks
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10, 12, 14])

## show the legend which was passed in the plot function as a lebel
plt.legend()

## save the figure (dpi 300 is good for , when I save the picture in high resolution)
plt.savefig('mygraph.png', dpi = 300)

plt.show()
```
