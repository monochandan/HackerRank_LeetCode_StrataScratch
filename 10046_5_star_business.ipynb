Find the top 5 states with the most 5 star businesses. Output the state name along with the number of 5-star businesses and order records by the number of 5-star businesses in descending order. In case there are ties in the number of businesses, return all the unique states. If two states have the same result, sort them in alphabetical order.

```python
df_business[df_business['stars'] == 5].groupby('state', as_index = False)['business_id'].count().rename(columns = {'business_id': 'n_businesses'}).sort_values(by = ['n_businesses','state'], ascending = [False,True ]).nlargest(5, 'n_businesses', 'all')
```

```python
# sql
# rank row-wise
# dense rank
# rank
with cte as (
select state, count(business_id) as n_businesses
from yelp_business
where start = 5
group by state
).
cte2 as (
select * , rank() over(order by n_businesses desc) as rnk
from cte )

select state, n_businesses
from cte2
where rnk <=5
order by n_businesses desc, state ## order by n_business, when there is a tie then alphabetical order by state
```
