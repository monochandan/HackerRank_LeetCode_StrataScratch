```python
df = pd.merge(orders ,customers, left_on = 'cust_id', right_on = 'id', how = 'left')
df['order_date'] = df['order_date'] # .dt.strftime('%Y-%m-%d')
#df
```

```python
df.groupby(['first_name', 'order_date'])['total_order_cost'].sum().reset_index().sort_values(by = 'total_order_cost', ascending =False).nlargest(1, 'total_order_cost', keep = 'all').rename(columns = {'total_order_cost': 'max_cost'})
#.head(1).rename(columns= {'total_order_cost': 'max_cost'})

```

```python
# sql
with cte as (
select c.first_name, sum(o.total_order_cost) AS total_order_cost,  date_format(o.order_date, '%Y-%m-%d') as order_date
from orders o
left join customers c
on c.id = o.cust_id
where date_format(o.order_date, '%Y-%m-%d') between '2019-02-01' and '2019-05-01'
group by c.first_name date_format(o.order_date, '%Y-%m-%d') ),

## ranking by total_order_cost column if there is multiple maximum value
cte2 as (
select *, rank() over(order by total_order_cost desc) as rnk
from cte )

select first_name, total_ordre_cost, order_date
from cte2
where rnk = 1
```
